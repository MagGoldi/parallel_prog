```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sizes = []
times = []

with open("files/result.txt", "r") as file:
    for line in file:
        parts = line.split()
        sizes.append(int(parts[1]))
        times.append(float(parts[3]))

plt.plot(sizes, times, marker='o')
plt.xlabel('Size')
plt.ylabel('Time, s')
plt.title('Time vs Size')
plt.grid()
plt.show()
```
